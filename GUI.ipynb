```python
import pandas
df = pandas.DataFrame(columns=['text','score'],index=[0,1,2])
df.loc[0] = pandas.Series({'text':"Awesome room service", 'score':1})
df.loc[1] = pandas.Series({'text':"Good view", 'score':0.5})
df.loc[2] = pandas.Series({'text':"Ok view", 'score':0})
df.loc[3] = pandas.Series({'text':"Bad room service", 'score':-0.5})
df.loc[4] = pandas.Series({'text':"Terrible prices", 'score':-1})
df.loc[5] = pandas.Series({'text':"Disastrous prices", 'score':-1})
df
```

```python
import tkinter
from tkinter import *

top = Tk()
top.title("Reddit Analyzer")

def listAll():
    print("all")
    t.delete(1.0,END)
    for (i,row) in df.iterrows():
        t.insert(END, row["text"] + '\n')

### add search functionality per se
def search():
    RB1.configure(state="normal")
    RB2.configure(state="normal")
    RB3.configure(state="normal")
    RB1.select()
    listAll()
    showKeywords()
    
def showKeywords():
    keyword1 = Label(top, text="room service", fg="blue", cursor="hand2")
    keyword1.grid(row=2,column=0)
    keyword1.bind("<Button-1>", lambda e: keywordContext(keyword1["text"]))

    keyword2 = Label(top, text="view", fg="blue", cursor="hand2")
    keyword2.grid(row=3,column=0)
    keyword2.bind("<Button-1>", lambda e: keywordContext(keyword2["text"]))

    keyword3 = Label(top, text="prices", fg="blue", cursor="hand2")
    keyword3.grid(row=4,column=0)
    keyword3.bind("<Button-1>", lambda e: keywordContext(keyword3["text"]))

def keywords():
    keywords = ["room service", "view", "prices"]
    for i in range(len(keywords)):
        keyword = keywords[i]
        link = Label(top, text=keyword, fg="blue", cursor="hand2")
        link.grid(row=2+i,column=0)
        print(keyword)
        link.bind("<Button-1>", lambda e: keywordContext(keyword))
    
def keywordContext(keyword):
    print("keyword context")
    child = tkinter.Toplevel(top)
    t_child = Text(child)
    t_child.pack()
    for (i,row) in df.iterrows():
        print(keyword,row["text"].lower())
        if keyword in row["text"].lower():
            t_child.insert(END, row["text"] + '\n')

def listPos():
    print("pos")
    t.delete(1.0,END)
    for (i,row) in df.iterrows():
        if row["score"] > 0:
            t.insert(END, row["text"] + '\n')
    
def listNeg():
    print("neg")
    t.delete(1.0,END)
    for (i,row) in df.iterrows():
        if row["score"] < 0:
            t.insert(END, row["text"] + '\n')

L1 = Label(top, text="Query: ")
L1.grid(row=0,column=0)

E1 = Entry(top, bd=5)
E1.bind('<Return>', lambda e: search())
E1.grid(row=0,column=1)
E1.focus_set()

submit = Button(top, text ="Search", command=search)
submit.grid(row=0,column=2)

choice = StringVar()
choice.set("all")
RB1 = Radiobutton(top, text="all", variable=choice, value="all", command=listAll, state="disabled")
RB1.grid(row=0,column=3)
RB2 = Radiobutton(top, text="pos", variable=choice, value="pos", command=listPos, state="disabled")
RB2.grid(row=0,column=4)
RB3 = Radiobutton(top, text="neg", variable=choice, value="neg", command=listNeg, state="disabled")
RB3.grid(row=0,column=5)

t = Text(top)
t.grid(row=1,column=0,columnspan=6)

top.mainloop()
```
